```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'C:\cs\data science\depi\house_file.csv'

df = pd.read_csv(url)
```

```python
print("dublicated data")
print(df.duplicated().sum())
```

```python
df.drop_duplicates()
```

```python
print(df.isna().sum())
df.info()
```

```python
#to remove sqrft and convert to numeric 
df["Size"] = df['Size'].str.replace("sqft","",regex=False)
df['Size'] = pd.to_numeric(df['Size'],errors="coerce")
```

```python
df['Taxes'] = df['Taxes'].str.replace("$","",regex=False)
df['Taxes'] = df['Taxes'].str.replace(",",".").astype(float)
df['Taxes']
```

```python
df['Price'] = df['Price'].str.replace("$","")
df['Price'] = df['Price'].str.replace(",","")
df['Price'] = df['Price'].replace({'X': None})
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
df['age'] = 2025 - df['date released']
df['age'] = df['age'].apply(lambda x: x if x >= 0 else 0)

df['age']
```

```python
df.isna().sum()
```

```python
rows_with_more_than_3_nulls = df[df.isnull().sum(axis=1) > 3]

print(rows_with_more_than_3_nulls)

df = df.dropna(thresh=df.shape[1] - 3)
```

```python
df.isna().sum()
```

```python
df.loc[:,'Size'] = df['Size'].fillna(df['Size'].median())
df.loc[:,'Price'] = df['Price'].fillna(df['Price'].median())
df.loc[:, 'date released'] = df['date released'].fillna(df['date released'].median())
```

```python
most_frequent = df['Type'].mode()[0]
df['Type'] = df['Type'].fillna(most_frequent)

most_frequent_addr = df['address'].mode()[0]
df['address'] = df['address'].fillna(most_frequent_addr)
```

```python
df["Taxes"] = df['Taxes'].fillna(df['Taxes'].median())
```
